```julia
function example(data;xcol,ycol,kwargs...)
    
    @vlplot(data=data,mark=:point, x=xcol, y=ycol,args...)
end
```

```julia
df |> vl"""
{ 
  "mark": "point",
  "width": 400,
  "height": 400,
  "encoding": {
    "x": {"field": "Horsepower", "type": "quantitative"},
    "y": {"field": "Miles_per_Gallon", "type": "quantitative"},
    "color": {
      "field": "Origin", "type":"nominal"
    }
  }
}
"""
```
