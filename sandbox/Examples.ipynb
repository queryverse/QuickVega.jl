```julia
using Revise, Pkg
Pkg.activate("../.")
using QuickVega, DataFrames, VegaLite, VegaDatasets, DataStructures
using Distributions, KernelDensity
using Contour
using Setfield
```

```julia
df = DataFrame(dataset("iris"));
```

```julia
QuickVega.setTheme!(:dark);
```

## Scatterplot 

```julia
s = scatterplot(df,:petalLength,:petalWidth,zcolor=:species,zsize=:petalLength,size_agg="mean",opacity=1.0,title="My Plot")
```

## Lineplot

```julia
x = float(collect(1:10))
y = float(x.^2)
s = lineplot(x,y,title="Line Plot")
```

## Barplot

```julia
x = rand(collect('a':'g'),20)
y = rand(20)

# Note that the theming is not working for multiple plots concatenated
p = [barplot(x,y,y_agg="mean",color="green",opacity=0.5) barplot(x,y,y_agg="max",color="purple",opacity=0.5) barplot(x,y,y_agg="min",color="red",opacity=0.5)]
```

```julia
p = barplot(df,:species,:petalWidth,zcolor=:petalLength,y_agg="mean",color_agg="mean",title="Bar Plot")
```

## Histogram

```julia
# Note that max_bins is a parameters of VegaLite, and does not match exactly the final number of bins in the plot
p = histogram(df,:petalLength,max_bins=20)
```

## Contour Plot

```julia
p = contourplot(df[!,:sepalLength],df[!,:sepalWidth],color_scheme="rainbow",title="Contour Plot with Lines")
```

```julia
p = contourplot(df[!,:sepalLength],df[!,:sepalWidth],color_scheme="plasma",mark="area", title="Contour Plot Filled")
```
